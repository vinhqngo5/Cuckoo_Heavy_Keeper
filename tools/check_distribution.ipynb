```python
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
from scipy import stats

def estimate_zipf_parameter(numbers, plot=False):
    """
    Estimate Zipf parameter using normalized frequencies
    """
    # Count frequencies and normalize
    counts = Counter(numbers)
    total = sum(counts.values())
    
    # Sort and normalize frequencies
    sorted_items = sorted(counts.items(), key=lambda x: x[1], reverse=True)
    frequencies = np.array([count for _, count in sorted_items])
    
    # Remove zero frequencies
    frequencies = frequencies[frequencies > 0]
    
    # Calculate ranks (1-based)
    ranks = np.arange(1, len(frequencies) + 1)
    
    # Log transform for fitting
    log_ranks = np.log(ranks)
    log_freqs = np.log(frequencies)
    
    # Linear regression
    slope, intercept, r_value, p_value, std_err = stats.linregress(log_ranks, log_freqs)
    slope = -0.7
    alpha = -slope
    
    if plot:
        plt.figure(figsize=(10, 6))
        plt.loglog(ranks, frequencies, 'b.', alpha=0.5, label='Observed')
        fitted = np.exp(intercept + slope * log_ranks)
        plt.loglog(ranks, fitted, 'r-', label=f'Zipf fit (α={alpha:.2f})')
        plt.xlabel('Rank')
        plt.ylabel('Normalized Frequency')
        plt.title('Zipf Distribution Fit (log-log scale)')
        plt.legend()
        plt.grid(True)
        plt.show()
        
        # Print goodness of fit
        print(f"R-squared: {r_value**2:.3f}")
    
    return alpha

if __name__ == "__main__":
    # Read data
    with open('/home/vinh/Q42024/VLDB_Revision_Cuckoo_Heavy_Keeper/data/TalkingData_AdTracking/2017_11_07_small', 'r') as f:
        numbers = [int(line.strip().split(",")[0]) for line in f if line.strip()]
    
    alpha = estimate_zipf_parameter(numbers, plot=True)
    print(f"Zipf parameter (α): {alpha:.3f}")
```
